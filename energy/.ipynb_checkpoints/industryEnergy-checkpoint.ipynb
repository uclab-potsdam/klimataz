```python
import pandas as pd
import numpy as np

def createWideFormat(df,column, values):
    #pivot table to wide format data
    orte = df["AGS"].unique()

    pivot_total = pd.DataFrame()

    for ort in orte:
            snippet = df[df["AGS"] == ort]

            temp = pd.pivot_table(snippet, index='Jahr',columns=column, values=values)
            temp = temp.reset_index(level=0)

            temp["ort_ags"] = ort
            temp["ort_name"]= str(snippet["1_Auspraegung_Label"].iloc[0]).lstrip()

            pivot_total = pd.concat([pivot_total, temp])
            
    return pivot_total

df = pd.read_csv("data/industry.csv",delimiter=";",encoding="latin1")

#prepare AGS and year for merging with population
df.rename({"1_Auspraegung_Code":"AGS","2_Auspraegung_Label":"Energietyp","Zeit":"Jahr","VBR001__Energieverbrauch_(einschl._nichtenergetischem_V.)__Tsd._MJ":"Energieverbrauch"},axis=1,inplace=True)

#to numeric
df["Energieverbrauch"] = df["Energieverbrauch"].replace({'-':np.NaN,".":np.NaN})
df["Energieverbrauch"] = pd.to_numeric(df["Energieverbrauch"])

#create wide format data
pivot = createWideFormat(df,"Energietyp","Energieverbrauch")
pivot["Anteil_Erneuerbar"] = pivot["Erneuerbare Energien"] / pivot["insgesamt"] * 100
pivot.to_csv("data/industry_wideformat.csv")
pivot

energies = ['Kohle', 'Heizöl', 'Erdgas', 'Erneuerbare Energien','Strom', 'Wärme', 'Sonstige Energieträger']

#melt dataset to long format data
melt = pd.melt(pivot,id_vars=["ort_ags","ort_name","Jahr"],value_vars = energies)
melt.to_csv("data/industry_longformat.csv")

df
```
