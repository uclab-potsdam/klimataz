```python
import pandas as pd

df = pd.read_csv("data/lak_primary_energy.csv")
#df.fillna(0,inplace=True)
print(df.columns)
df.drop(columns=["Anteil Erneuerbarer Energieträger an der Strom erzeugung","Anteil Erneuerbarer Energieträger am Brutto stromver brauch"],inplace=True)
df.rename({'Anteil Erneuerbarer Energieträger am Primär energie verbrauch':"share_renewables_primary_energy_consumption"},axis=1,inplace=True)

df.to_csv("data/lak_primary_energy_longformat.csv")
df

pivot = df.pivot(index="Land",columns="Jahr",values="share_renewables_primary_energy_consumption").reset_index(level=0)
pivot.to_csv("data/lak_primary_energy_wideformat.csv")
pivot
```

```python
df2 = pd.read_csv("data/lak_primary_renewables.csv",delimiter=";",encoding='latin1')

df2.Land.replace({"Baden-Wï¿½rttemberg":"Baden-Württemberg",'Thï¿½ringen':"Thüringen"},inplace=True)
df2.rename({"Klï¿½rgas, Deponiegas":"Klärgas, Deponiegas"},axis=1,inplace=True)

df2.drop(columns=["Anm","Stand"],inplace=True)

#merge with non-renewable forms
df_total = pd.read_csv("data/lak_primary_total.csv",delimiter=";",encoding='latin1')

df_total.Land.replace({"Baden-Wï¿½rttemberg":"Baden-Württemberg",'Thï¿½ringen':"Thüringen"},inplace=True)
df_total.rename({"Mineralï¿½le und Mineralï¿½l produkte":"Mineralöle und Mineralöl produkte",
"Erneuerbare Energietrï¿½ger":"Gesamt Erneuerbare Energieträger","Andere Energietrï¿½ger":"Andere Energieträger"},axis=1,inplace=True)

df_total.drop(columns=["Anm","Stand"],inplace=True)

df3 = pd.merge(df_total,df2,on=["Jahr","Land"],how="outer")
df3

#calculate values for renewables energies
print(df3['Gesamt Erneuerbare Energieträger'].equals(df3['Insgesamt_y']))

df3["Gesamt Nichterneuerbar"] = df3["Insgesamt_x"]-df3["Gesamt Erneuerbare Energieträger"]
df3["Anteil Erneuerbar"] = df3["Gesamt Erneuerbare Energieträger"] / df3["Insgesamt_x"] * 100

df3.drop(columns=['Insgesamt_y'],inplace=True)
df3

#all energies
print(df3.columns)
energies = ['Steinkohle', 'Braunkohle',
       'Mineralöle und Mineralöl produkte', 'Gase',
       'Gesamt Erneuerbare Energieträger', 'Stromaustauschsaldo',
       'Kernenergie', 'Andere Energieträger', 'Klärgas, Deponiegas',
       'Wasserkraft', 'Windkraft', 'Solarenergie', 'Biomasse 1)',
       'Sonstige erneuerbare Energien 2)']

#melt dataset to long format data
melt = pd.melt(df3,id_vars=["Land","Jahr"],value_vars = energies)
melt.to_csv("data/lak_primary_detailed_longformat.csv")

#pivot table to wide format data
pivot_total = pd.DataFrame()

for energy in energies:
        temp = pd.pivot_table(df3, index='Land',columns="Jahr", values=energy)
        temp = temp.reset_index(level=0)

        temp["energy"] = energy
        pivot_total = pd.concat([pivot_total, temp])

pivot_total.to_csv("data/lak_primary_detailed_wideformat.csv")

df3
```
