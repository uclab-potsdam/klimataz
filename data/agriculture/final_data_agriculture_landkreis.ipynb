```python
# import necessary libraries

import pandas as pd
import numpy as np
```

```python
#load dataset from local file
source_df = pd.read_csv("data/41141-03-01-4_flat.csv",delimiter=";", encoding='latin1', low_memory=False)

#remove subgroups of animals, Milchkühe and Zuchtsauen
source_df = source_df[(source_df['2_Auspraegung_Code'] != 'TIER-111') & (source_df['2_Auspraegung_Code'] != 'TIER-22')]

source_df
```

```python
tiere_df = source_df.copy()

#extract year from date
tiere_df['Jahr']=tiere_df['Zeit'].str[-4:]

#drop unnecessary columns
tiere_df.drop(columns=['Zeit','Statistik_Code', 'Statistik_Label', 'Zeit_Code', 'Zeit_Label', 
       '1_Merkmal_Code', '1_Merkmal_Label', 
       '2_Merkmal_Code', '2_Merkmal_Label',
       '2_Auspraegung_Code',
       'BTR034__Landwirtschaftliche_Betriebe_mit_Viehhaltung__Anzahl',
       'BTR038__Landwirtschaftliche_Betriebe_mit_100_u.m._GV__Anzahl',
       'VIE003__Viehbestand__GV', 'BTR034__Betriebe__Anzahl'],inplace=True)

#rename columns
tiere_df.rename(columns={'1_Auspraegung_Code':'AGS','1_Auspraegung_Label':'Name','2_Auspraegung_Label':'Tier','TIE003__Tiere__Anzahl':'Anzahl'},inplace=True)

#replace '-' with 0 and '.' with 'NaN according to Zeichnerklärung
tiere_df['Anzahl'].replace('-','0',inplace=True)
tiere_df['Anzahl'] = tiere_df['Anzahl'].apply(pd.to_numeric,errors='coerce')

tiere_df['AGS']=tiere_df['AGS'].replace("DG",0)

tiere_df
```

```python
#pivot dataframe to show animals and ears as columns
pt_df = tiere_df.pivot(index=['AGS','Name'],columns=['Tier','Jahr'],values=['Anzahl'])
pt_df
```

```python
#collapse levels of column names and join energy type to year
pt_df.columns = pt_df.columns.map('_'.join)

#sort columns so indicators are together
pt_df.columns=sorted(pt_df.columns)

#prepend all columns with indicator name
pt_df=pt_df.add_prefix('agriculture_animalcount_')

#convert AGS to numeric and set as new index
pt_df.reset_index(inplace=True)
pt_df['AGS'] = pd.to_numeric(pt_df['AGS'])
pt_df.set_index('AGS',inplace=True)
pt_df.sort_index(ascending=True,inplace=True)

#strip whitespace from region names
pt_df['Name']=pt_df['Name'].str.strip()

pt_df
```

```python
#save dataframe to file
pt_df.to_csv("data/final_data_landkreis_agriculture_animalcount.csv")
```
