```python
# import necessary libraries

import pandas as pd
import numpy as np
```

```python
#load dataset from local file
source_df = pd.read_csv("data/41141-04-02-4_flat.csv",delimiter=";", encoding='latin1', low_memory=False)
source_df
```

```python
#load 2016 dataset from local file to complete Viehbestand data for Rheinland Pfalz
vb16_df = pd.read_csv("data/41141-03-01-4-2016.csv",delimiter=";", encoding='latin1', low_memory=False)

#keep only viehbestand column and AGS for merge
vb16_df = vb16_df[['AGS','Viehbestand 2016']]

#rename columns according to convention
vb16_df.columns = ['AGS','Tierzahl_GV_2016']

#check if data for RheinlandPfalz is complete
vb16_df[(vb16_df['AGS']>=7000)&(vb16_df['AGS']<8000)]
```

```python
#load 2016 dataset from local file to complete Landwirtschaftsfläche data for Rheinland Pfalz
lf16_df = pd.read_csv("data/41141-04-02-4-2016.csv",delimiter=";", encoding='latin1', low_memory=False)

#keep only landwirtschaftlich genutzte fläche data and AGS for merge
lf16_df = lf16_df[['AGS','Landwirtschaftlich genutzte Flaeche 2016']]

#rename columns according to convention
lf16_df.columns = ['AGS','Landwirtschaft_Flaeche_2016']

#check if data for RheinlandPfalz is complete
lf16_df[(lf16_df['AGS']>=7000)&(lf16_df['AGS']<8000)]
```

```python
source_df.columns
```

```python
#keep only relevant columns
gv_df = source_df[['1_Auspraegung_Code','1_Auspraegung_Label','FLC017__Landwirtschaftlich_genutzte_Flaeche__ha','VIE003__Viehbestand__GV']].copy()

#rename columns
gv_df.rename(columns={'1_Auspraegung_Code':'AGS','1_Auspraegung_Label':'Name','FLC017__Landwirtschaftlich_genutzte_Flaeche__ha':'Landwirtschaft_Flaeche','VIE003__Viehbestand__GV':'Tierzahl_GV'},inplace=True)

#rename AGS for Deutschland from DG to 0 according to project convention
gv_df['AGS']=gv_df['AGS'].replace("DG",0)

#convert AGS column to numeric for merge
gv_df['AGS'] = gv_df['AGS'].apply(pd.to_numeric,errors='coerce')

#merge viehbestand and landwirtschaftsfläche data for 2016 on AGS
gv_df = gv_df.merge(vb16_df,on='AGS', how='left')
gv_df = gv_df.merge(lf16_df,on='AGS', how='left')

#sort columns to facilitate forward fill
gv_df = gv_df[['AGS', 'Name', 'Landwirtschaft_Flaeche_2016','Landwirtschaft_Flaeche', 'Tierzahl_GV_2016','Tierzahl_GV']]

#check Rheinland Pfalz data
gv_df[(gv_df['AGS']>=7000)&(gv_df['AGS']<8000)]
```

```python
#replace '-' with 0 and '.' with 'NaN according to Zeichnerklärung
gv_df.replace('-','0',inplace=True)

#pd.to_numeric with 'coerce' replaces the only non-numeric values '.' with NaN
gv_df['Tierzahl_GV'] = gv_df['Tierzahl_GV'].apply(pd.to_numeric,errors='coerce')
gv_df['Landwirtschaft_Flaeche'] = gv_df['Landwirtschaft_Flaeche'].apply(pd.to_numeric,errors='coerce')
gv_df['Tierzahl_GV_2016'] = gv_df['Tierzahl_GV_2016'].apply(pd.to_numeric,errors='coerce')
gv_df['Landwirtschaft_Flaeche_2016'] = gv_df['Landwirtschaft_Flaeche_2016'].apply(pd.to_numeric,errors='coerce')

gv_df[(gv_df['AGS']>=7000)&(gv_df['AGS']<8000)]
```

```python
#replace landwirtschaft and tierzahl columns with forward filled columns
gv_df[['Landwirtschaft_Flaeche_2016','Landwirtschaft_Flaeche']]= gv_df[['Landwirtschaft_Flaeche_2016','Landwirtschaft_Flaeche']].fillna(method='ffill',axis=1)
gv_df[['Tierzahl_GV_2016','Tierzahl_GV']]= gv_df[['Tierzahl_GV_2016','Tierzahl_GV']].fillna(method='ffill',axis=1)

gv_df[(gv_df['AGS']>=7000)&(gv_df['AGS']<8000)]
```

```python
#drop columns with 2016 data
gv_df = gv_df[['AGS', 'Name', 'Landwirtschaft_Flaeche','Tierzahl_GV']]

#calculate animals per area for ranking
gv_df['Tiere_pro_Flaeche'] = gv_df['Tierzahl_GV']/gv_df['Landwirtschaft_Flaeche']
gv_df
```

## Landkreise

```python
#dictionary to assign thirds based on rank
bins=[1,133,266,399]
names=['besten Drittel','mittleren Drittel','schlechtesten Drittel']
```

```python
#load list of current landkreise
kreise_akt = pd.read_csv("../general_data/gerics_landkreis_names_for_ranking.csv")
kreise_akt = kreise_akt[['AGS','Gerics_Name','Gerics_Bundesland']].copy()

kreise_liste = kreise_akt['AGS'].to_list()

kreise_akt
```

```python
agr_kreise = gv_df.copy()
agr_kreise
```

```python
agr_filter = agr_kreise[agr_kreise['AGS'].isin(kreise_liste)].copy().reset_index(drop=True)
agr_filter
```

```python
#check to see if data on any landkreise missing
missing = [x for x in kreise_liste if x not in agr_kreise['AGS'].to_list()]
missing
```

```python
#load gerics names and replace
agr_filter= agr_filter.merge(kreise_akt, on='AGS',how='left')
agr_filter
```

```python
agr_tiere = agr_filter[['AGS','Gerics_Name','Gerics_Bundesland','Tierzahl_GV','Landwirtschaft_Flaeche','Tiere_pro_Flaeche']].copy()
agr_tiere.sort_values('Tiere_pro_Flaeche',ascending=True,inplace=True)
agr_tiere.reset_index(drop=True,inplace=True)

agr_tiere
```

```python
#check which landkreise are still NaN
agr_tiere[pd.isna(agr_tiere['Tiere_pro_Flaeche'])]

#only urban landkreise in which number of animals are "unknown" or "confidential"
#cross checked that NaN DOES NOT imply zero here, because some have Tierzahl info for organic farms in these Landkreise
```

```python
#what do we do with NaNs for postcard text and ranking?
```

```python
agr_tiere['rank'] = agr_tiere['Tiere_pro_Flaeche'].rank(na_option = 'keep')
agr_tiere['third'] = pd.cut(agr_tiere['rank'], bins, labels=names,include_lowest=True)
agr_tiere
```

```python
agr_final = agr_tiere[['AGS','Gerics_Name','Gerics_Bundesland','Landwirtschaft_Flaeche','Tierzahl_GV','Tiere_pro_Flaeche','rank','third']].copy()
agr_final.columns=['AGS','Name','Bundesland','Landwirtschaft_Flaeche','Tierzahl_GV','GV_pro_Flaeche','agriculture_rank','agriculture_third']
agr_final
```

## Bundesländer

```python
#dictionary to assign thirds based on rank
bl_bins=[1,5,11,16]
bl_names=['besten Drittel','mittleren Drittel','schlechtesten Drittel']
```

```python
blagr_df = gv_df[gv_df['AGS']<17].copy().reset_index(drop=True)
blagr_df['Name']=blagr_df['Name'].str.strip()
blagr_df.replace('Baden-Wï¿½rttemberg, Land','Baden-Württemberg',inplace=True)
blagr_df.replace('Thï¿½ringen','Thüringen',inplace=True)
blagr_df
```

```python
dl_tiere = blagr_df.loc[0]
blagr_tiere = blagr_df[blagr_df['AGS']>0].copy()

blagr_tiere.sort_values('Tiere_pro_Flaeche',inplace=True)
blagr_tiere.reset_index(drop=True,inplace=True)
blagr_tiere
```

```python
blagr_tiere['rank'] = blagr_tiere['Tiere_pro_Flaeche'].rank(na_option = 'keep')
blagr_tiere['third'] = pd.cut(blagr_tiere['rank'], bl_bins, labels=bl_names,include_lowest=True)
blagr_tiere['Bundesland'] = blagr_tiere['Name']
blagr_tiere
```

```python
dl_tiere['rank'] = 0
dl_tiere['third'] = ""
dl_tiere['Bundesland']="Deutschland"
dl_tiere
```

```python
blagr_tiere=blagr_tiere.append(dl_tiere)
blagr_tiere
```

```python
blagr_final = blagr_tiere[['AGS','Name','Bundesland','Landwirtschaft_Flaeche','Tierzahl_GV','Tiere_pro_Flaeche','rank','third']].copy()
blagr_final.columns=['AGS','Name','Bundesland','Landwirtschaft_Flaeche','Tierzahl_GV','GV_pro_Flaeche','agriculture_rank','agriculture_third']
blagr_final
```

## 3. Merge both datasets and sort by AGS

```python
complete_df = pd.concat([blagr_final,agr_final], ignore_index=True).copy()
final_df = complete_df.sort_values('AGS').reset_index(drop=True)
final_df
```

## 4. Load Bundesland level data and merge

```python
#load dataset from local file
blag_df = pd.read_csv("data/final_data_bundesland_agriculture_animalcount.csv")
blag_df
```

```python
blag_df.columns
```

```python
blag_short = blag_df[['Name','agriculture_animalcount_bydensity_Insgesamt_2020','agriculture_animalcount_bydensity_davon_auf_unter_2_0_GV_ha_2020','agriculture_animalcount_bydensity_davon_Ökolandbau_2020']].copy()
blag_short
```

```python
blag_short['Anteil_unter_2GV_pro_Hektar'] = blag_short['agriculture_animalcount_bydensity_davon_auf_unter_2_0_GV_ha_2020']/blag_short['agriculture_animalcount_bydensity_Insgesamt_2020']*100
blag_short['Anteil_oekolandbau'] = blag_short['agriculture_animalcount_bydensity_davon_Ökolandbau_2020']/blag_short['agriculture_animalcount_bydensity_Insgesamt_2020']*100
blag_short.drop(columns=['agriculture_animalcount_bydensity_Insgesamt_2020','agriculture_animalcount_bydensity_davon_auf_unter_2_0_GV_ha_2020','agriculture_animalcount_bydensity_davon_Ökolandbau_2020'],inplace=True)
blag_short.columns = ['Bundesland','Anteil_unter_2GV_pro_Hektar','Anteil_oekolandbau']
blag_short
```

```python
final_merged = final_df.merge(blag_short,on="Bundesland",how="left")
final_merged
```
