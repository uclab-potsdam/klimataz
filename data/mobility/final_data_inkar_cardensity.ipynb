```python
#inkar.de, download "Pkw-Dichte"-Dataset 1995-2019
```

```python
# import necessary libraries

import pandas as pd
```

```python
df = pd.read_csv("data/2207_inkar_pkwdichte_1995_2019.csv",delimiter=";",header=1,decimal=",")
df
```

```python
cd_df = df.copy('deep')

#drop unnecessary column
cd_df.drop(columns='Unnamed: 2',inplace=True)

#add indicator prefix
cd_df=cd_df.add_prefix('PKW_Dichte_')

#rename ags and name columns
cd_df.rename(columns={'PKW_Dichte_Unnamed: 0':'AGS','PKW_Dichte_Unnamed: 1': 'Name'},inplace=True)

#set ags as index
cd_df.set_index('AGS',inplace=True)

cd_df
```

```python
#save dataframe to file
cd_df.to_csv("data/final_data_landkreis_mobility_inkar_pkwdichte.csv")
```
