```python
#load final data file for mobility indicator
#and use most current data to create ranking
#document whether the landkreis belongs to the top, middle or bottom third
```

```python
# import necessary libraries
import numpy as np
import pandas as pd
```

### 1. Landkreise

```python
#dictionary to assign thirds based on rank
bins=[1,133,266,399]
names=['besten Drittel','mittleren Drittel','schlechtesten Drittel']
```

```python
#load list of current landkreise
kreise_akt = pd.read_csv("../general_data/gerics_landkreis_names_for_ranking.csv")
kreise_akt = kreise_akt[['AGS','Gerics_Name','Gerics_Bundesland']].copy()

kreise_liste = kreise_akt['AGS'].to_list()

kreise_akt
```

```python
#load dataset from local file
mor_kreise = pd.read_csv("data/final_data_landkreis_mobility_fueltype.csv")
mor_kreise.head(20)
```

```python
mor_filter = mor_kreise[mor_kreise['AGS'].isin(kreise_liste)].copy().reset_index(drop=True)
mor_filter
```

```python
#check to see if data on any landkreise missing
missing = [x for x in kreise_liste if x not in mor_kreise['AGS'].to_list()]
missing
```

```python
#load gerics names and replace
mor_filter= mor_filter.merge(kreise_akt, on='AGS',how='left').drop(columns='Name')
mor_filter
```

```python
mor_filter.columns
```

```python
mor_density = mor_filter[['AGS','Gerics_Name','Gerics_Bundesland','mobility_fueltype_cardensity_Gesamt_2020','mobility_fueltype_cardensity_Gesamt_2021','mobility_fueltype_cardensity_Elektro_2020','mobility_fueltype_cardensity_Elektro_2021','mobility_fueltype_cardensity_Hybrid_2020','mobility_fueltype_cardensity_Hybrid_2021']].copy()
mor_density= mor_density.fillna(method='ffill',axis=1,limit=1)
mor_density['car_density'] = mor_density['mobility_fueltype_cardensity_Gesamt_2021']/10
mor_density['ev_hybrid'] = (mor_density['mobility_fueltype_cardensity_Elektro_2021']+mor_density['mobility_fueltype_cardensity_Hybrid_2021'])/10
mor_density.sort_values('car_density',ascending=True,inplace=True)
mor_density.reset_index(drop=True,inplace=True)
mor_density
```

```python
mor_density['rank'] = mor_density.index+1
mor_density['third'] = pd.cut(mor_density['rank'], bins, labels=names,include_lowest=True)
mor_density
```

```python
mor_final = mor_density[['AGS','Gerics_Name','Gerics_Bundesland','car_density','ev_hybrid','rank','third']].copy()
mor_final.columns=['AGS','Name','Bundesland','mobility_car_density','mobility_ev_hybrid','mobility_rank','mobility_third']
mor_final
```

```python
#load bundesland data and merge und bundesland name
blmo_df = pd.read_csv("data/final_data_bundesland_mobility_mid.csv")
blmo_short = blmo_df[['Name']].copy()
blmo_short['mobility_pc_km_umweltfreundlich_2002'] = blmo_df[['mobility_mid_wegkm_percentage_Fahrrad_2002','mobility_mid_wegkm_percentage_Fuß_2002','mobility_mid_wegkm_percentage_ÖPV_2002']].sum(axis=1)
blmo_short['mobility_pc_km_umweltfreundlich_2017'] = blmo_df[['mobility_mid_wegkm_percentage_Fahrrad_2017','mobility_mid_wegkm_percentage_Fuß_2017','mobility_mid_wegkm_percentage_ÖPV_2017']].sum(axis=1)
#blmo_short['mobility_pc_km_Auto_2017'] = blmo_df[['mobility_mid_wegkm_percentage_MIV_(Fahrer)_2017','mobility_mid_wegkm_percentage_MIV_(Mitfahrer)_2017']].sum(axis=1)
blmo_short.columns=['Bundesland','mobility_pc_km_umweltfreundlich_2002','mobility_pc_km_umweltfreundlich_2017']
blmo_short
```

```python
mor_merged = mor_final.merge(blmo_short,on='Bundesland',how='left')
mor_merged
```

## 2. Bundesländer

```python
#dictionary to assign thirds based on rank
bl_bins=[1,5,11,16]
bl_names=['besten Drittel','mittleren Drittel','schlechtesten Drittel']
```

```python
blmor_filter = mor_kreise[mor_kreise['AGS']<17].copy()
blmor_filter.replace('Baden-Württemberg, Land','Baden-Württemberg',inplace=True)
blmor_filter
```

```python
blmor_density = blmor_filter[['AGS','Name','mobility_fueltype_cardensity_Gesamt_2020','mobility_fueltype_cardensity_Gesamt_2021','mobility_fueltype_cardensity_Elektro_2020','mobility_fueltype_cardensity_Elektro_2021','mobility_fueltype_cardensity_Hybrid_2020','mobility_fueltype_cardensity_Hybrid_2021']].copy()
blmor_density= blmor_density.fillna(method='ffill',axis=1,limit=1)
blmor_density['car_density'] = blmor_density['mobility_fueltype_cardensity_Gesamt_2021']/10
blmor_density['ev_hybrid'] = (blmor_density['mobility_fueltype_cardensity_Elektro_2021']+blmor_density['mobility_fueltype_cardensity_Hybrid_2021'])/10

dl_density = blmor_density.loc[0]
blmor_density = blmor_density[blmor_density['AGS']>0]

blmor_density.sort_values('car_density',ascending=True,inplace=True)
blmor_density.reset_index(drop=True,inplace=True)
blmor_density
```

```python
blmor_density['rank'] = blmor_density.index+1
blmor_density['third'] = pd.cut(blmor_density['rank'], bins=bl_bins, labels=bl_names,include_lowest=True)
blmor_density['Bundesland'] = blmor_density['Name']
blmor_density
```

```python
dl_density['rank'] = 0
dl_density['third'] = ""
dl_density['Bundesland']="Deutschland"
dl_density
```

```python
blmor_density=blmor_density.append(dl_density)
blmor_density
```

```python
blmor_final = blmor_density[['AGS','Name','Bundesland','car_density','ev_hybrid','rank','third']].copy()
blmor_final.columns=['AGS','Name','Bundesland','mobility_car_density','mobility_ev_hybrid','mobility_rank','mobility_third']
blmor_final
```

```python
blmor_merged = blmor_final.merge(blmo_short,on='Bundesland',how='left')
blmor_merged
```

## 3. Merge both datasets and sort by AGS

```python
complete_df = pd.concat([blmor_merged,mor_merged], ignore_index=True).copy()
final_df = complete_df.sort_values('AGS').reset_index(drop=True)
final_df
```

```python
#save dataframe to file
final_df.to_csv("data/final_postcard_data_mobility.csv")
```
