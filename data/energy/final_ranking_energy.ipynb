```python
# import necessary libraries
import numpy as np
import pandas as pd
```

## 1. Landkreise

```python
#dictionary to assign landkreise to thirds based on rank
bins=[1,133,266,399]
names=['besten Drittel','mittleren Drittel','schlechtesten Drittel']
```

```python
#dictionary to assign bundesland to thirds based on rank
bl_bins=[1,5,11,16]
bl_names=['besten Drittel','mittleren Drittel','schlechtesten Drittel']
```

```python
#load list of current landkreise
kreise_akt = pd.read_csv("../general_data/gerics_landkreis_names_for_ranking.csv")
kreise_akt = kreise_akt[['AGS','Gerics_Name','Gerics_Bundesland']].copy()

kreise_liste = kreise_akt['AGS'].to_list()

kreise_akt
```

```python
#load dataset from local file
enr_kreise = pd.read_csv("data/final_data_landkreis_energy_industryenergy.csv")
enr_kreise.head(20)
```

```python
enr_filter = enr_kreise[enr_kreise['AGS'].isin(kreise_liste)].copy().reset_index(drop=True)
enr_filter
```

```python
#check to see if data on any landkreise missing
missing = [x for x in kreise_liste if x not in enr_kreise['AGS'].to_list()]
missing
```

```python
#load gerics names and replace
enr_filter= enr_filter.merge(kreise_akt, on='AGS',how='left').drop(columns='Name')
enr_filter
```

```python
print(enr_filter.columns)
```

```python
column_list = [x for x in enr_filter.columns if '_2020' in x if '_Anteil_' in x]
column_list.extend(['AGS','Gerics_Name',
 'Gerics_Bundesland'])
column_list
```

```python
enr_energien = enr_filter[column_list].copy()
#enr_energien['Insgesamt'] = enr_energien['energy_industry_consumption_insgesamt_2020']
#enr_energien['Erneuerbare'] = enr_energien[ 'energy_industry_consumption_Erneuerbare Energien_2020']
#enr_energien['Fossil'] = enr_energien[['energy_industry_consumption_Heizöl_2020','energy_industry_consumption_Kohle_2020']].sum(axis=1)
#enr_energien['Andere'] = enr_energien[['energy_industry_consumption_Sonstige Energieträger_2020','energy_industry_consumption_Strom_2020','energy_industry_consumption_Wärme_2020']].sum(axis=1)
#enr_energien['Geheim'] = enr_energien['energy_industry_consumption_insgesamt_2020']-enr_energien[['Fossil','Andere','Erneuerbare']].sum(axis=1)
enr_energien['Erneuerbare_Anteil'] = enr_energien['energy_industry_consumption_Anteil_Erneuerbar_2020'].astype(float)
#enr_energien['Fossil_Anteil'] = enr_energien['Fossil']*100/enr_energien['Insgesamt']
#enr_energien['Andere_Anteil'] = enr_energien['Andere']*100/enr_energien['Insgesamt']
enr_energien['Geheim_Anteil'] = enr_energien['energy_industry_consumption_Anteil_Geheim_2020'].astype(float)
enr_energien['Erneuerbare_Anteil'].fillna(0,inplace=True)
enr_energien['Geheim_Anteil'].fillna(100,inplace=True)

enr_energien =  enr_energien[['AGS', 'Gerics_Name','Gerics_Bundesland','Erneuerbare_Anteil','Geheim_Anteil']]
enr_energien.sort_values('Erneuerbare_Anteil',ascending=False,inplace=True)
#enr_energien.reset_index(drop=True,inplace=True)
enr_energien
```

```python
#enr_energien['energy_rank'] = enr_energien.index+1
#enr_energien['energy_third'] = pd.cut(enr_energien['rank'], bins, labels=names,include_lowest=True)
#enr_energien[enr_energien['Geheim_Anteil']==100]
```

```python
enr_final = enr_energien.copy()
enr_final.columns=['AGS', 'Name','Bundesland','energy_Erneuerbare_Anteil','energy_Geheim_Anteil']
enr_final
```

```python
#load bundesland data and merge on bundesland name
blen_df = pd.read_csv("data/final_data_bundesland_energy_primaryconsumption.csv")
blen_df = blen_df.fillna(method='ffill',axis=1)
blen_short = blen_df[['Name','energy_primaryconsumption_Anteil Erneuerbar_2020']].copy()
blen_short.columns=['Bundesland','energy_Anteil_Erneuerbar_Bundesland']
blen_short['energy_rank'] = blen_short['energy_Anteil_Erneuerbar_Bundesland'].rank(ascending=False)
blen_short['energy_third'] = pd.cut(blen_short['energy_rank'], bl_bins, labels=bl_names,include_lowest=True)

blen_short
```

```python
enr_merged = enr_final.merge(blen_short,on='Bundesland',how='left')
enr_merged
```

## 2. Bundesländer

```python
blenr_filter = enr_kreise[enr_kreise['AGS']<17].copy()
blenr_filter.replace('Baden-Württemberg, Land','Baden-Württemberg',inplace=True)
blenr_filter['Bundesland'] = blenr_filter['Name']
blenr_filter
```

```python
blcolumn_list = [x for x in enr_filter.columns if '_2020' in x if '_Anteil_' in x]
blcolumn_list.extend(['AGS','Name',
 'Bundesland'])
blcolumn_list
```

```python
blenr_energien = blenr_filter[blcolumn_list].copy()
#blenr_energien['Insgesamt'] = blenr_energien['energy_industry_consumption_insgesamt_2020']
#blenr_energien['Erneuerbare'] = blenr_energien[ 'energy_industry_consumption_Erneuerbare Energien_2020']
#blenr_energien['Fossil'] = blenr_energien[['energy_industry_consumption_Heizöl_2020','energy_industry_consumption_Kohle_2020']].sum(axis=1)
#blenr_energien['Andere'] = blenr_energien[['energy_industry_consumption_Sonstige Energieträger_2020','energy_industry_consumption_Strom_2020','energy_industry_consumption_Wärme_2020']].sum(axis=1)
#blenr_energien['Geheim'] = blenr_energien['energy_industry_consumption_insgesamt_2020']-blenr_energien[['Fossil','Andere','Erneuerbare']].sum(axis=1)
blenr_energien['Erneuerbare_Anteil'] = blenr_energien['energy_industry_consumption_Anteil_Erneuerbar_2020'].astype(float)
#blenr_energien['Fossil_Anteil'] = blenr_energien['Fossil']*100/blenr_energien['Insgesamt']
#blenr_energien['Andere_Anteil'] = blenr_energien['Andere']*100/blenr_energien['Insgesamt']
blenr_energien['Geheim_Anteil'] = blenr_energien['energy_industry_consumption_Anteil_Geheim_2020'].astype(float)
blenr_energien['Erneuerbare_Anteil'].fillna(0,inplace=True)
blenr_energien['Geheim_Anteil'].fillna(100,inplace=True)
blenr_energien =  blenr_energien[['AGS', 'Name','Bundesland','Erneuerbare_Anteil','Geheim_Anteil']]

dl_energien = blenr_energien.loc[0]
blenr_energien = blenr_energien[blenr_energien['AGS']>0]

blenr_energien.sort_values('Erneuerbare_Anteil',ascending=False,inplace=True)
blenr_energien
```

```python
blenr_energien=blenr_energien.merge(blen_short,on='Bundesland',how='left')
blenr_energien
```

```python
#load bundesland data and merge und bundesland name
dlen_df = pd.read_csv("data/final_data_deutschland_energy_primaryconsumption.csv")
dlen_short = dlen_df[['Name','energy_primaryconsumption_Anteil Erneuerbar_2020']].copy()
dlen_short.columns=['Name','energy_Anteil_Erneuerbar_Bundesland']
dlen_short
```

```python
dl_energien['energy_rank'] = ""
dl_energien['energy_third'] = ""
dl_energien['energy_Anteil_Erneuerbar_Bundesland'] = dlen_short['energy_Anteil_Erneuerbar_Bundesland'][0]
dl_energien
```

```python
blenr_energien=blenr_energien.append(dl_energien)
blenr_energien
```

```python
blenr_final = blenr_energien.copy()
blenr_final.columns=['AGS', 'Name','Bundesland','energy_Erneuerbare_Anteil','energy_Geheim_Anteil', 'energy_Anteil_Erneuerbar_Bundesland','energy_rank','energy_third']
blenr_final
```

## 3. Merge both datasets and sort by AGS

```python
complete_df = pd.concat([blenr_final,enr_merged], ignore_index=True).copy()
final_df = complete_df.sort_values('AGS').reset_index(drop=True)
final_df['energy_Geheim_Anteil'] = pd.to_numeric(final_df['energy_Geheim_Anteil'])
final_df['energy_Anteil_Erneuerbar_Bundesland'] = pd.to_numeric(final_df['energy_Anteil_Erneuerbar_Bundesland'])

final_df
```

```python
#save dataframe to file
final_df.to_csv("data/final_postcard_data_energy.csv")
```
