```python
#import necessary libraries

import pandas as pd
```

```python
#list of regions

regions = {"Deutschland":0,
           "Baden-Württemberg":8,
           "Bayern":9,"Berlin":11,
           "Brandenburg":12,
           "Bremen":4,
           "Hamburg":2,
           "Hessen":6,
           "Mecklenburg-Vorpommern":13,
           "Niedersachsen":3,
           "Nordrhein-Westfalen":5,
           "Rheinland-Pfalz":7,"Saarland":10,"Sachsen":14,"Sachsen-Anhalt":15,"Schleswig-Holstein":1,"Thüringen":16}

regions
```

```python
#read renewables data from csv
de_df = pd.read_csv("data/deutschland_erneuerbare.csv",delimiter=",")

#clean up labels
de_df.rename({"Klärgas,Deponiegas":"Klärgas, Deponiegas"},axis=1,inplace=True)

#drop unnecessary columns
de_df.drop(columns=["Zeilenname","Zeile"],inplace=True)

de_df
```

```python
#read non-renewable energies from csv
dee_df = pd.read_csv("data/deutschland_primaerenergie.csv",delimiter=",")

#clean up labels
dee_df.rename({"Sonstige Energieträger":"Andere Energieträger","Mineralöle und Mineralöl Produkte":"Mineralöle und Mineralöl produkte"},axis=1,inplace=True)

#drop unnecessary columns
dee_df.drop(columns=["Energiebilanz der Bundesrepublik 2020","Zeile"],inplace=True)

dee_df
```

```python
#merge renewable and non-renewable energy dataframes
merged_df = pd.merge(de_df,dee_df,on=["Jahr"],how="outer")
merged_df
```

```python
#test values
print(merged_df['Gesamt Erneuerbare Energieträger'].equals(merged_df['Summe_x']))
```

```python
#Sums are not equal due to small rounding errors in original data
```

```python
#calculate aggregate values for renewable and non-renewable energies
merged_df["Gesamt Nichterneuerbar"] = merged_df["Summe_y"]-merged_df["Gesamt Erneuerbare Energieträger"]
merged_df["Anteil Erneuerbar"] = merged_df["Gesamt Erneuerbare Energieträger"] / merged_df["Summe_y"] * 100

#rename totals column
merged_df.drop(columns=['Summe_x'],inplace=True)
merged_df.rename(columns={'Summe_y':'Insgesamt'},inplace=True)

#change year column to string to facilitate renaming columns after pivot
merged_df['Jahr'] = merged_df['Jahr'].astype(str)

merged_df
```

```python
merged_df['Land']='Deutschland'
merged_df.columns
```

```python
#pivot table to have regions in rows and years and energy type in columns
final_df = merged_df.pivot(index = 'Land', columns = ['Jahr'], values= ['Insgesamt', 'Steinkohle', 'Braunkohle',
       'Mineralöle und Mineralöl produkte', 'Gase',
       'Gesamt Erneuerbare Energieträger', 'Stromaustauschsaldo',
       'Kernenergie', 'Andere Energieträger', 'Klärgas, Deponiegas',
       'Wasserkraft', 'Windkraft', 'Solarenergie', 'Biomasse',
       'Sonstige erneuerbare Energien', 'Gesamt Nichterneuerbar',
       'Anteil Erneuerbar'])

#collapse levels of column names and join energy type to year
final_df.columns = final_df.columns.map('_'.join)

#prepend all columns with indicator name
final_df=final_df.add_prefix('energy_primaryconsumption_')

final_df
```

```python
#add AGS ids from the regions dictionary using the region names
final_df['AGS']=final_df.index.map(regions)

#set AGS as index and copy region names to first column
first_column=final_df.index
final_df.set_index('AGS', inplace=True)
final_df.insert(0,'Name',first_column)

final_df
```

```python
#save dataframe to file
final_df.to_csv("data/final_data_deutschland_energy_primaryconsumption.csv")
```
