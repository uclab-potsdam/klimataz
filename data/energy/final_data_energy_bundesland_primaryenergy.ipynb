```python
#import necessary libraries

import pandas as pd
```

```python
#list of regions

regions = {"Deutschland":0,
           "Baden-Württemberg":8,
           "Bayern":9,"Berlin":11,
           "Brandenburg":12,
           "Bremen":4,
           "Hamburg":2,
           "Hessen":6,
           "Mecklenburg-Vorpommern":13,
           "Niedersachsen":3,
           "Nordrhein-Westfalen":5,
           "Rheinland-Pfalz":7,"Saarland":10,"Sachsen":14,"Sachsen-Anhalt":15,"Schleswig-Holstein":1,"Thüringen":16}

regions
```

```python
#read renewables data from csv
be_df = pd.read_csv("data/lak_primary_renewables.csv",delimiter=";",encoding='latin1')

#clean up labels
be_df.Land.replace({"Baden-Wï¿½rttemberg":"Baden-Württemberg",'Thï¿½ringen':"Thüringen"},inplace=True)
be_df.rename({"Klï¿½rgas, Deponiegas":"Klärgas, Deponiegas"},axis=1,inplace=True)

#drop unnecessary columns
be_df.drop(columns=["Anm","Stand"],inplace=True)

be_df
```

```python
#read non-renewable energies from csv
et_df = pd.read_csv("data/lak_primary_total.csv",delimiter=";",encoding='latin1')

#clean up labels
et_df.Land.replace({"Baden-Wï¿½rttemberg":"Baden-Württemberg",'Thï¿½ringen':"Thüringen"},inplace=True)
et_df.rename({"Mineralï¿½le und Mineralï¿½l produkte":"Mineralöle und Mineralöl produkte",
"Erneuerbare Energietrï¿½ger":"Gesamt Erneuerbare Energieträger","Andere Energietrï¿½ger":"Andere Energieträger"},axis=1,inplace=True)

#drop unnecessary columns
et_df.drop(columns=["Anm","Stand"],inplace=True)

et_df
```

```python
#merge renewable and non-renewable energy dataframes
merged_df = pd.merge(et_df,be_df,on=["Jahr","Land"],how="outer")
merged_df
```

```python
#test values
print(merged_df['Gesamt Erneuerbare Energieträger'].equals(merged_df['Insgesamt_y']))
```

```python
#calculate aggregate values for renewable and non-renewable energies
merged_df["Gesamt Nichterneuerbar"] = merged_df["Insgesamt_x"]-merged_df["Gesamt Erneuerbare Energieträger"]
merged_df["Anteil Erneuerbar"] = merged_df["Gesamt Erneuerbare Energieträger"] / merged_df["Insgesamt_x"] * 100

#rename totals column
merged_df.drop(columns=['Insgesamt_y'],inplace=True)
merged_df.rename(columns={'Insgesamt_x':'Insgesamt'},inplace=True)

#change year column to string to facilitate renaming columns after pivot
merged_df['Jahr'] = merged_df['Jahr'].astype(str)

merged_df.columns
```

```python
#pivot table to have regions in rows and years and energy type in columns
final_df = merged_df.pivot(index = 'Land', columns = ['Jahr'], values= ['Insgesamt', 'Steinkohle', 'Braunkohle',
       'Mineralöle und Mineralöl produkte', 'Gase',
       'Gesamt Erneuerbare Energieträger', 'Stromaustauschsaldo',
       'Kernenergie', 'Andere Energieträger', 'Klärgas, Deponiegas',
       'Wasserkraft', 'Windkraft', 'Solarenergie', 'Biomasse 1)',
       'Sonstige erneuerbare Energien 2)', 'Gesamt Nichterneuerbar',
       'Anteil Erneuerbar'])

#collapse levels of column names and join energy type to year
final_df.columns = final_df.columns.map('_'.join)

#prepend all columns with indicator name
final_df=final_df.add_prefix('energy_primaryconsumption_')

final_df
```

```python
#add AGS ids from the regions dictionary using the region names
final_df['AGS']=final_df.index.map(regions)

#set AGS as index and copy region names to first column
first_column=final_df.index
final_df.set_index('AGS', inplace=True)
final_df.insert(0,'Name',first_column)

#sort the dataframe by AGS id
final_df.sort_index(inplace=True)

#unit of measure
final_df["Unit"] = "TJoule"

final_df
```

```python
#save dataframe to file
final_df.to_csv("data/final_data_bundesland_energy_primaryconsumption.csv")
```
