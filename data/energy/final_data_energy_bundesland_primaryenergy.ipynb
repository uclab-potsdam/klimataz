```python
#import necessary libraries

import pandas as pd
```

```python
#list of regions

regions = {"Deutschland":0,
           "Baden-Württemberg":8,
           "Bayern":9,"Berlin":11,
           "Brandenburg":12,
           "Bremen":4,
           "Hamburg":2,
           "Hessen":6,
           "Mecklenburg-Vorpommern":13,
           "Niedersachsen":3,
           "Nordrhein-Westfalen":5,
           "Rheinland-Pfalz":7,"Saarland":10,"Sachsen":14,"Sachsen-Anhalt":15,"Schleswig-Holstein":1,"Thüringen":16}

regions
```

```python
#read renewables data from csv
be_df = pd.read_csv("data/lak_primary_renewables.csv",delimiter=";",encoding='latin1')

#clean up labels
be_df.Land.replace({"Baden-Wï¿½rttemberg":"Baden-Württemberg",'Thï¿½ringen':"Thüringen"},inplace=True)
be_df.rename({"Klï¿½rgas, Deponiegas":"Klärgas, Deponiegas"},axis=1,inplace=True)

#drop unnecessary columns
be_df.drop(columns=["Anm","Stand"],inplace=True)

be_df
```

```python
#read non-renewable energies from csv
et_df = pd.read_csv("data/lak_primary_total.csv",delimiter=";",encoding='latin1')

#clean up labels
et_df.Land.replace({"Baden-Wï¿½rttemberg":"Baden-Württemberg",'Thï¿½ringen':"Thüringen"},inplace=True)
et_df.rename({"Mineralï¿½le und Mineralï¿½l produkte":"Mineralöle und Mineralöl produkte",
"Erneuerbare Energietrï¿½ger":"Gesamt Erneuerbare Energieträger","Andere Energietrï¿½ger":"Andere Energieträger"},axis=1,inplace=True)

#drop unnecessary columns
et_df.drop(columns=["Anm","Stand"],inplace=True)

et_df
```

```python
#https://ag-energiebilanzen.de/daten-und-fakten/auswertungstabellen/
#download excel file, go to table 2.1, show more decimals, export table 2.1 as csv

#german data
germany = pd.read_csv("data/germany_primaryenergy_rawdata.csv")

#reformat data
germany = germany.loc[0:14].copy()
germany = germany.transpose().reset_index().set_index(2)
germany = germany.rename(columns=germany.iloc[1]).drop(germany.index[1])
germany.drop(germany.columns.difference(['Steinkohle','Braunkohle','Mineralöle','Gase','Erneuerbare Energien','Außenhandelssaldo Strom','Sonstige Energieträger','Kernenergie','Insgesamt']), 1, inplace=True)
germany.index = pd.to_numeric(germany.index, errors='coerce')
germany = germany.loc[1990:2020].copy()
#germany[germany.columns] = df[cols].apply(pd.to_numeric, errors='coerce')

#convert from petajoule to terajoule
germany = germany.applymap(pd.to_numeric, errors='coerce')
germany = germany.applymap(lambda x:x*1000)

germany = germany.reset_index(level=0)
germany.rename({"Mineralöle":"Mineralöle und Mineralöl produkte","Außenhandelssaldo Strom":"Stromaustauschsaldo",
"Erneuerbare Energien":"Gesamt Erneuerbare Energieträger","Sonstige Energieträger":"Andere Energieträger",2:"Jahr"},axis=1,inplace=True)
germany["Land"] = "Deutschland"
germany
```

```python
#merge renewable and non-renewable energy dataframes
#merged_df = et_df.merge(be_df,on=["Jahr","Land"],how="outer").merge(germany,on=["Jahr","Land"],how="outer")
#merged_df = pd.merge(et_df,be_df,on=["Jahr","Land"],how="outer").
merged_df = pd.concat([et_df,germany])

#test values
#print(merged_df['Gesamt Erneuerbare Energieträger'].equals(merged_df['Insgesamt_y']))

merged_df
```

```python
#calculate aggregate values for renewable and non-renewable energies
merged_df["Gesamt Nichterneuerbar"] = merged_df["Insgesamt"]-merged_df["Gesamt Erneuerbare Energieträger"]
merged_df["Anteil Erneuerbar"] = merged_df["Gesamt Erneuerbare Energieträger"] / merged_df["Insgesamt"] * 100

#rename totals column
#merged_df.drop(columns=['Insgesamt_y'],inplace=True)
merged_df.rename(columns={'Biomasse 1)':'Biomasse','Sonstige erneuerbare Energien 2)':'Sonstige erneuerbare Energien'},inplace=True)

#change year column to string to facilitate renaming columns after pivot

merged_df['Jahr'] = merged_df['Jahr'].astype(int).astype(str)

merged_df.columns

#interpolate missing values
merged_df = merged_df.groupby('Land').apply(lambda group: group.interpolate(axis=0))
merged_df
```

```python
#pivot table to have regions in rows and years and energy type in columns
final_df = merged_df.pivot(index = 'Land', columns = ['Jahr'], values= ['Insgesamt', 'Steinkohle', 'Braunkohle',
       'Mineralöle und Mineralöl produkte', 'Gase',
       'Gesamt Erneuerbare Energieträger', 'Stromaustauschsaldo',
       'Kernenergie', 'Andere Energieträger','Gesamt Nichterneuerbar',
       'Anteil Erneuerbar'])


#collapse levels of column names and join energy type to year
final_df.columns = final_df.columns.map('_'.join)

#prepend all columns with indicator name
final_df=final_df.add_prefix('energy_primaryconsumption_')

final_df
```

```python
#add AGS ids from the regions dictionary using the region names
final_df['AGS']=final_df.index.map(regions)

#copy region names to first column
first_column=final_df.index
final_df.insert(0,'Name',first_column)

#set AGS as index
final_df.set_index('AGS', inplace=True)

#sort the dataframe by AGS id
final_df.sort_index(inplace=True)

#unit of measure
final_df["Unit"] = "TJoule"

final_df
```

```python
#save dataframe to file
final_df.to_csv("data/final_data_bundesland_energy_primaryconsumption.csv")
```
