```python
#load final data file for building indicator landkreise
#and use most current data to create ranking
#document whether the landkreis belongs to the top, middle or bottom third
```

```python
# import necessary libraries
import numpy as np
import pandas as pd
```

```python
#dictionary to assign thirds based on rank
bins=[1,134,268,401]
names=['oberen Drittel','mittleren Drittel','unteren Drittel']
```

```python
#load list of current landkreise
kreise_akt = pd.read_csv("../general_data/kreise_geocoordinates.csv")
kreise_liste = kreise_akt['AGS'].to_list()

#correct values for Hamburg and Berlin
kreise_liste.extend([2,11])
kreise_liste.remove(2000)
kreise_liste.remove(11000)

kreise_liste
```

```python
#load dataset from local file
nbr_kreise = pd.read_csv("data/final_data_landkreis_newbuildings_renewables.csv")
nbr_kreise
```

```python
nbr_filter = nbr_kreise[nbr_kreise['AGS'].isin(kreise_liste)].copy().reset_index(drop=True)
nbr_filter
```

```python
#check to see if data on any landkreise missing
missing = [x for x in kreise_liste if x not in nbr_kreise['AGS'].to_list()]
missing
```

```python
nbr_filter.columns
```

```python
nbr_erneuerbare = nbr_filter[['AGS','Name','buildings_newbuilding_energy_renewables_2020','buildings_newbuilding_energy_fossils_2020']].copy()
nbr_erneuerbare['anteil_erneuerbare'] = (nbr_erneuerbare['buildings_newbuilding_energy_renewables_2020']*100)/(nbr_erneuerbare['buildings_newbuilding_energy_renewables_2020']+nbr_erneuerbare['buildings_newbuilding_energy_fossils_2020'])
nbr_erneuerbare.sort_values('anteil_erneuerbare',ascending=False,inplace=True)
nbr_erneuerbare.reset_index(drop=True,inplace=True)
nbr_erneuerbare
```

```python
nbr_erneuerbare['rank'] = nbr_erneuerbare.index+1
nbr_erneuerbare['third'] = pd.cut(nbr_erneuerbare['rank'], bins, labels=names,include_lowest=True)
nbr_erneuerbare
```

```python
nbr_erneuerbare['postcard_text'] = nbr_erneuerbare.agg(lambda x:f"In {x['Name']} wurden 2020 {round(x['anteil_erneuerbare'])} Prozent der neuen Gebäude mit erneuerbaren Energien geheizt. Damit liegt der Landkreis im Vergleich im {x['third']}." ,axis=1)
nbr_erneuerbare['postcard_text'][136]
```

```python
nbr_final = nbr_erneuerbare[['AGS','Name','anteil_erneuerbare','rank','third','postcard_text']]
nbr_final
```

```python
#save dataframe to file
nbr_final.to_csv("data/final_rank_landkreis_newbuildings_renewables.csv")
```
