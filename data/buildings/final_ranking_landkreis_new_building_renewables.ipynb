```python
# import necessary libraries
import numpy as np
import pandas as pd
```

### 1. Landkreise

```python
#dictionary to assign thirds based on rank
bins=[1,133,266,399]
names=['besten Drittel','mittleren Drittel','schlechtesten Drittel']
```

```python
#load list of current landkreise
kreise_akt = pd.read_csv("../general_data/gerics_landkreis_names_for_ranking.csv")
kreise_akt = kreise_akt[['AGS','Gerics_Name','Gerics_Bundesland']].copy()
#kreise_akt.replace(to_replace=2000,value=2,inplace=True)
#kreise_akt.replace(to_replace=11000,value=11,inplace=True)

kreise_liste = kreise_akt['AGS'].to_list()

#correct values for Hamburg and Berlin
#kreise_liste.extend([2,11])
#kreise_liste.remove(2000)
#kreise_liste.remove(11000)

#kreise_liste
kreise_akt
```

```python
#load dataset from local file
nbr_kreise = pd.read_csv("data/final_data_landkreis_newbuildings_renewables.csv")
nbr_kreise.head(20)
```

```python
nbr_filter = nbr_kreise[nbr_kreise['AGS'].isin(kreise_liste)].copy().reset_index(drop=True)
nbr_filter
```

```python
#load gerics names and replace
nbr_filter= nbr_filter.merge(kreise_akt, on='AGS',how='left').drop(columns='Name')
nbr_filter
```

```python
#check to see if data on any landkreise missing
missing = [x for x in kreise_liste if x not in nbr_kreise['AGS'].to_list()]
missing
```

```python
nbr_filter.columns
```

```python
nbr_erneuerbare = nbr_filter[['AGS','Gerics_Name','Gerics_Bundesland','buildings_newbuilding_energy_renewables_2016','buildings_newbuilding_energy_fossils_2016','buildings_newbuilding_energy_renewables_2020','buildings_newbuilding_energy_fossils_2020']].copy()
nbr_erneuerbare['anteil_erneuerbare_2016'] = (nbr_erneuerbare['buildings_newbuilding_energy_renewables_2016']*100)/(nbr_erneuerbare['buildings_newbuilding_energy_renewables_2016']+nbr_erneuerbare['buildings_newbuilding_energy_fossils_2016'])
nbr_erneuerbare['anteil_erneuerbare_2020'] = (nbr_erneuerbare['buildings_newbuilding_energy_renewables_2020']*100)/(nbr_erneuerbare['buildings_newbuilding_energy_renewables_2020']+nbr_erneuerbare['buildings_newbuilding_energy_fossils_2020'])
nbr_erneuerbare.sort_values('anteil_erneuerbare_2020',ascending=False,inplace=True)
nbr_erneuerbare.reset_index(drop=True,inplace=True)
nbr_erneuerbare
```

```python
nbr_erneuerbare['rank'] = nbr_erneuerbare.index+1
nbr_erneuerbare['third'] = pd.cut(nbr_erneuerbare['rank'], bins, labels=names,include_lowest=True)
nbr_erneuerbare
```

```python
nbr_erneuerbare['postcard_text'] = nbr_erneuerbare.agg(lambda x:f"In {x['Gerics_Name']} wurden 2020 {round(x['anteil_erneuerbare_2020'])} Prozent der neuen Gebäude mit erneuerbaren Energien geheizt, 2016 waren es {round(x['anteil_erneuerbare_2016'])} Prozent. Damit liegt der Landkreis im Vergleich im {x['third']}." ,axis=1)
nbr_erneuerbare['postcard_text'][136]
```

```python
nbr_erneuerbare.columns
```

```python
nbr_final = nbr_erneuerbare[['AGS','Gerics_Name','Gerics_Bundesland','anteil_erneuerbare_2016','anteil_erneuerbare_2020','rank','third']].copy()
nbr_final.columns=['AGS','Name','Bundesland','buildings_anteil_erneuerbare_2016','buildings_anteil_erneuerbare_2020','buildings_rank','buildings_third']
nbr_final
```

```python
#load bundesland data and merge und bundesland name
bl_df = pd.read_csv("data/final_data_bundesland_heating.csv")
bl_short = bl_df[['Name','buildings_heating_2002','buildings_heating_2020']].copy()
bl_short.columns=['Bundesland','buildings_heating_2002','buildings_heating_2020']
bl_short
```

```python
nbr_merged = nbr_final.merge(bl_short,on='Bundesland',how='left')
nbr_merged
```

### 2. Bundesländer

```python
#dictionary to assign thirds based on rank
bl_bins=[1,5,11,16]
bl_names=['besten Drittel','mittleren Drittel','schlechtesten Drittel']
```

```python
blr_filter = nbr_kreise[nbr_kreise['AGS']<17].copy()
blr_filter.replace('Baden-Württemberg, Land','Baden-Württemberg',inplace=True)
blr_filter
```

```python
blr_erneuerbare = blr_filter[['AGS','Name','buildings_newbuilding_energy_renewables_2016','buildings_newbuilding_energy_fossils_2016','buildings_newbuilding_energy_renewables_2020','buildings_newbuilding_energy_fossils_2020']].copy()
blr_erneuerbare['anteil_erneuerbare_2016'] = (blr_erneuerbare['buildings_newbuilding_energy_renewables_2016']*100)/(blr_erneuerbare['buildings_newbuilding_energy_renewables_2016']+blr_erneuerbare['buildings_newbuilding_energy_fossils_2016'])
blr_erneuerbare['anteil_erneuerbare_2020'] = (blr_erneuerbare['buildings_newbuilding_energy_renewables_2020']*100)/(blr_erneuerbare['buildings_newbuilding_energy_renewables_2020']+blr_erneuerbare['buildings_newbuilding_energy_fossils_2020'])
dl_erneuerbare = blr_erneuerbare.loc[0]
blr_erneuerbare = blr_erneuerbare[blr_erneuerbare['AGS']>0]
blr_erneuerbare
```

```python
blr_erneuerbare.sort_values('anteil_erneuerbare_2020',ascending=False,inplace=True)
blr_erneuerbare.reset_index(drop=True,inplace=True)
blr_erneuerbare
```

```python
blr_erneuerbare['rank'] = blr_erneuerbare.index+1
blr_erneuerbare['third'] = pd.cut(blr_erneuerbare['rank'], bl_bins, labels=bl_names,include_lowest=True)
blr_erneuerbare['Bundesland']=blr_erneuerbare['Name']
blr_erneuerbare
```

```python
dl_erneuerbare['rank'] = 0
dl_erneuerbare['third'] = ""
dl_erneuerbare['Bundesland']="Deutschland"
dl_erneuerbare
```

```python
blr_erneuerbare=blr_erneuerbare.append(dl_erneuerbare)
blr_erneuerbare
```

```python
blr_final = blr_erneuerbare[['AGS','Name','Bundesland','anteil_erneuerbare_2016','anteil_erneuerbare_2020','rank','third']].copy()
blr_final.columns=['AGS','Name','Bundesland','buildings_anteil_erneuerbare_2016','buildings_anteil_erneuerbare_2020','buildings_rank','buildings_third']
blr_final
```

```python
blr_merged = blr_final.merge(bl_short,on='Bundesland',how='left')
blr_merged
```
