```python
#import necessary libraries
import pandas as pd
```

```python
#import gerics landkreis data and standardize columns
gerics_landkreis = pd.read_csv('other/gerics/gerics_original_data/final_gerics_rcp45_2036_2065_heisze_tage_median.csv')
gerics_landkreis.drop(columns=['Unnamed: 0'],inplace=True)
gerics_landkreis.columns=['AGS', 'Name', 'Bundesland', '1971-2000','2036-2065', 'percentage change']
gerics_landkreis =  gerics_landkreis[(gerics_landkreis['AGS'] != 11000) & (gerics_landkreis['AGS'] != 2000)]
gerics_landkreis
```

```python
#import gerics bundesland data and standardize columns
gerics_bundesland = pd.read_csv('other/gerics/gerics_original_data/final_gerics_rcp45_2036_2065_heisze_tage_median_de_bundesland.csv')
gerics_bundesland.drop(columns=['Unnamed: 0','Median','Postcard Intro'],inplace=True)
gerics_bundesland['Bundesland'] = gerics_bundesland['Name']
gerics_bundesland
```

```python
#merge both datasets into one dataframe
gerics_df = gerics_landkreis.append(gerics_bundesland).sort_values('AGS').copy()
gerics_df
```

```python
#create base dataframe for postcard texts
postcards_df = gerics_df[['AGS','Name']].copy()
postcards_df['gerics_text'] = gerics_df.agg(lambda x: f'''Hallo!\n\nIch schicke dir Grüße aus dem schönen <span className='location-name'>{x['Name']}</span>. Derzeit wird es hier an durchschnittlich <span className="gerics-data">{str(round(x['1971-2000'],1)).replace(".", ",")} Tagen</span> im Jahr über 30 Grad, aber wenn wir so weitermachen, könnten es Mitte des Jahrhunderts schon <span className="gerics-data">{str(round(x['2036-2065'],1)).replace(".", ",")}</span> sein. <span className="gerics-data">{round(x['percentage change'])} Prozent mehr!</span>\n\n''',
                                            axis=1)
postcards_df
```
