```python
# import necessary libraries

import pandas as pd
import numpy as np
```

```python
#load dataset from local file
source_df = pd.read_csv("data/32121-01-02-4_flat_haushaltsabfaelle.csv",delimiter=";", encoding='latin1', low_memory=False)

# #remove subgroups of waste (everything except for organic waste total) Haus- und Sperrmüll, Getrennt erfasste organische Abfälle, Getrennt erfasste Wertstoffe, Sonstige Abfälle, Biotonne, abbaubare Garten- und Parkabfälle and Insgesamt
# source_df = source_df[(source_df['2_Auspraegung_Code'] != 'ABFALLART100') & (source_df['2_Auspraegung_Code'] != 'ABFALLART300') & (source_df['2_Auspraegung_Code'] != 'ABFALLART400') & (source_df['2_Auspraegung_Code'] != 'ABFALLART201') & (source_df['2_Auspraegung_Code'] != 'ABFALLART202') & (source_df['2_Auspraegung_Label'] != "Insgesamt")]

#remove subgroups of waste Haus- und Sperrmüll, Getrennt erfasste organische Abfälle, Getrennt erfasste Wertstoffe, Sonstige Abfälle and Insgesamt
source_df = source_df[(source_df['2_Auspraegung_Code'] != 'ABFALLART100') & (source_df['2_Auspraegung_Code'] != 'ABFALLART300') & (source_df['2_Auspraegung_Code'] != 'ABFALLART400') & (source_df['2_Auspraegung_Label'] != "Insgesamt") ]

source_df
```

```python
waste_df = source_df.copy()

#extract year from date
waste_df['Jahr']=waste_df['Zeit'].astype(str)

#drop unnecessary columns
waste_df.drop(columns=['Zeit','Statistik_Code', 'Statistik_Label', 'Zeit_Code', 'Zeit_Label', 
       '1_Merkmal_Code', '1_Merkmal_Label', 
       '2_Merkmal_Code', '2_Merkmal_Label',
       '2_Auspraegung_Code'],inplace=True)

#rename columns
waste_df.rename(columns={'1_Auspraegung_Code':'AGS','1_Auspraegung_Label':'Name','2_Auspraegung_Label':'Abfall','AEW009__Aufkommen_an_Haushaltsabfaellen_(o.E-altgeraete)__t':'weight'},inplace=True)

#replace '-' with 0 and '.' with 'NaN according to Zeichnerklärung
waste_df['weight'].replace('-','0',inplace=True)
waste_df['weight'] = waste_df['weight'].apply(pd.to_numeric,errors='coerce')

waste_df['AGS']=waste_df['AGS'].replace("DG",0)

waste_df
```

```python
#pivot dataframe to show waste and years as columns
pt_df = waste_df.pivot(index=['AGS','Name'],columns=['Abfall','Jahr'],values=['weight'])

#rename columns
pt_df.rename(columns={'Getrennt erfasste organische Abfälle':'sum'},inplace=True)

#rename columns sub groups organic waste
pt_df.rename(columns={'biolog. abbaubare Garten- und Parkabfälle':'gartenPark', 'Abfälle aus der Biotonne':'biotonne'},inplace=True)

pt_df
```

```python
#collapse levels of column names and join waste type to year
pt_df.columns = pt_df.columns.map('_'.join)

#sort columns so indicators are together
pt_df=pt_df[sorted(pt_df.columns)]

#prepend all columns with indicator name
pt_df=pt_df.add_prefix('waste_organicwaste_')

#convert AGS to numeric and set as new index
pt_df.reset_index(inplace=True)
pt_df['AGS'] = pd.to_numeric(pt_df['AGS'])
pt_df.set_index('AGS',inplace=True)
pt_df.sort_index(ascending=True,inplace=True)

#strip whitespace from region names
pt_df['Name']=pt_df['Name'].str.strip()

#unit of measure
pt_df["Unit"] = "Kg per person"


pt_df
#sorted(pt_df.columns)
```

```python
#save dataframe to file
pt_df.to_csv("data/final_data_landkreis_waste_organicwaste.csv")
```

## Normalize waste_df by population

```python
#copy waste_df
norm_wa = waste_df.copy()

#set variable types for merge
norm_wa['AGS'] = norm_wa['AGS'].astype(int)
norm_wa['Jahr'] = norm_wa['Jahr'].astype(int)
norm_wa
```

```python
# load population data from general data
pop_df = pd.read_csv("../general_data/population.csv",delimiter=";", encoding='latin1', low_memory=False)

#correct Germany AGS
pop_df['AGS'] = pop_df['1_Auspraegung_Code'].astype('string').str.replace('DG','0')

#extract year from time column
pop_df['Jahr'] = pop_df['Zeit'].astype('string').str[-4:]

#select only necessary coluns and rename
pop_df = pop_df[['AGS','Jahr','BEVSTD__Bevoelkerungsstand__Anzahl']]
pop_df.columns=['AGS','Jahr','Bevölkerung']

#set variable types for merge
pop_df['AGS'] = pop_df['AGS'].astype(int)
pop_df['Jahr'] = pop_df['Jahr'].astype(int)

pop_df
```

```python
#merge waste_df with population
merged_wa = norm_wa.merge(pop_df, on = ['AGS','Jahr'], how='left')
merged_wa['Bevölkerung'] =merged_wa['Bevölkerung'].apply(pd.to_numeric, errors='coerce')
merged_wa
```

```python
merged_wa['kg_per_person'] = merged_wa['weight']/merged_wa['Bevölkerung']*1000
merged_wa['Name'] = merged_wa['Name'].str.strip()
merged_wa
```

```python
#convert year to string to facilitate collapsing column levels
merged_wa['Jahr']=merged_wa['Jahr'].astype(str)

#pivot dataframe to show waste and years as columns
nw_df = merged_wa.pivot(index=['AGS','Name'],columns=['Abfall','Jahr'],values=['kg_per_person'])

#rename columns
nw_df.rename(columns={'Getrennt erfasste organische Abfälle':'sum'},inplace=True)

#rename columns sub groups organic waste
nw_df.rename(columns={'biolog. abbaubare Garten- und Parkabfälle':'gartenPark', 'Abfälle aus der Biotonne':'biotonne'},inplace=True)

nw_df
```

```python
#collapse levels of column names and join waste type to year
nw_df.columns = nw_df.columns.map('_'.join)

#sort columns so indicators are together
nw_df=nw_df[sorted(nw_df.columns)]

#prepend all columns with indicator name
nw_df=nw_df.add_prefix('waste_organicwaste_')

#convert AGS to numeric and set as new index
nw_df.reset_index(inplace=True)
nw_df['AGS'] = pd.to_numeric(nw_df['AGS'])
nw_df.set_index('AGS',inplace=True)
nw_df.sort_index(ascending=True,inplace=True)

#strip whitespace from region names
nw_df['Name']=nw_df['Name'].str.strip()

#unit of measure
nw_df["Unit"] = "Kg per person"

nw_df
```

```python
#save dataframe to file
nw_df.to_csv("data/final_data_landkreis_waste_organicwaste_normalized.csv")
```
