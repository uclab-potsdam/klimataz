```python
#load final data file for waste indicator
#and use most current data to create ranking
#document whether the landkreis belongs to the top, middle or bottom third
```

```python
# import necessary libraries
import numpy as np
import pandas as pd
```

```python
#dictionary to assign thirds based on rank
bins=[1,133,266,399]
names=['besten Drittel','mittleren Drittel','schlechtesten Drittel']
```

```python
#load list of current landkreise
kreise_akt = pd.read_csv("../general_data/gerics_landkreis_names_for_ranking.csv")
kreise_akt = kreise_akt[['AGS','Gerics_Name','Gerics_Bundesland']].copy()

kreise_liste = kreise_akt['AGS'].to_list()

kreise_akt
```

```python
#load dataset from local file
war_kreise = pd.read_csv("data/final_data_landkreis_waste_organicwaste_normalized.csv")
war_kreise.head(20)
```

```python
war_filter = war_kreise[war_kreise['AGS'].isin(kreise_liste)].copy().reset_index(drop=True)
war_filter
```

```python
#check to see if data on any landkreise missing
missing = [x for x in kreise_liste if x not in war_kreise['AGS'].to_list()]
missing
```

```python
#load gerics names and replace
war_filter= war_filter.merge(kreise_akt, on='AGS',how='left')
war_filter
```

```python
war_filter.isna().sum()
```

```python
war_organic = war_filter[['AGS','Gerics_Name','Gerics_Bundesland','waste_organicwaste_kg_per_person_sum_2016','waste_organicwaste_kg_per_person_sum_2020']].copy()
war_organic.sort_values('waste_organicwaste_kg_per_person_sum_2020',ascending=False,inplace=True)
war_organic.reset_index(drop=True,inplace=True)
war_organic
```

```python
war_organic['rank'] = war_organic['waste_organicwaste_kg_per_person_sum_2020'].rank(na_option = 'keep',ascending=False)
war_organic['third'] = pd.cut(war_organic['rank'], bins, labels=names,include_lowest=True)
war_organic
```

```python
war_final = war_organic[['AGS','Gerics_Name','Gerics_Bundesland','waste_organicwaste_kg_per_person_sum_2016','waste_organicwaste_kg_per_person_sum_2020','rank','third']].copy()
war_final.columns=['AGS','Name','Bundesland','kg_per_person_2016','kg_per_person_2020','waste_rank','waste_third']
war_final
```

## 2. Bundesländer

```python
#dictionary to assign thirds based on rank
bl_bins=[1,5,11,16]
bl_names=['besten Drittel','mittleren Drittel','schlechtesten Drittel']
```

```python
blwar_filter = war_kreise[war_kreise['AGS']<17].copy()
blwar_filter.replace('Baden-Württemberg, Land','Baden-Württemberg',inplace=True)
blwar_filter
```

```python
blwar_organic = blwar_filter[['AGS','Name','waste_organicwaste_kg_per_person_sum_2016','waste_organicwaste_kg_per_person_sum_2020']].copy()

dl_organic = blwar_organic.loc[0]
blwar_organic = blwar_organic[blwar_organic['AGS']>0]

blwar_organic.sort_values('waste_organicwaste_kg_per_person_sum_2020',ascending=False,inplace=True)
blwar_organic.reset_index(drop=True,inplace=True)
blwar_organic
```

```python
blwar_organic['rank'] = blwar_organic['waste_organicwaste_kg_per_person_sum_2020'].rank(na_option = 'keep',ascending=False)
blwar_organic['third'] = pd.cut(blwar_organic['rank'], bl_bins, labels=bl_names,include_lowest=True)
blwar_organic['Bundesland'] = blwar_organic['Name']
blwar_organic
```

```python
dl_organic['rank'] = 0
dl_organic['third'] = ""
dl_organic['Bundesland']="Deutschland"
dl_organic
```

```python
blwar_organic=blwar_organic.append(dl_organic)
blwar_organic
```

```python
blwar_final = blwar_organic[['AGS','Name','Bundesland','waste_organicwaste_kg_per_person_sum_2016','waste_organicwaste_kg_per_person_sum_2020','rank','third']].copy()
blwar_final.columns=['AGS','Name','Bundesland','kg_per_person_2016','kg_per_person_2020','waste_rank','waste_third']
blwar_final
```

## 3. Merge both datasets and sort by AGS

```python
complete_df = pd.concat([blwar_final,war_final], ignore_index=True).copy()
final_df = complete_df.sort_values('AGS').reset_index(drop=True)
final_df
```

```python
#save dataframe to file
final_df.to_csv("data/final_postcard_data_waste.csv")
```
