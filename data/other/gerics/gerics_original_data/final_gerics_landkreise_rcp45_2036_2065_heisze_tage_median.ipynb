```python
import pandas as pd
```

```python
#list of regions

regions = {"Deutschland":0,
           "Baden-Württemberg":8,
           "Bayern":9,"Berlin":11,
           "Brandenburg":12,
           "Bremen":4,
           "Hamburg":2,
           "Hessen":6,
           "Mecklenburg-Vorpommern":13,
           "Niedersachsen":3,
           "Nordrhein-Westfalen":5,
           "Rheinland-Pfalz":7,"Saarland":10,"Sachsen":14,"Sachsen-Anhalt":15,"Schleswig-Holstein":1,"Thüringen":16}

regions = dict([(value, key) for key, value in regions.items()])

regions
```

```python
df = pd.read_csv("GERICS_Klimaausblick_Landkreise_rcp45_2036-2065.txt")
df
```

```python
#filter out only Heiße Tage
gerics_df = df.copy()
heisz_df = gerics_df[gerics_df['Klimakennwert'] == 'Heiße Tage'].reset_index()
heisz_df = heisz_df[['Klimakennwert','Landkreis Name','Landkreis Kennziffer (Kreisschlüssel)','Median der Änderungen','Einheit','HYRAS Beobachtungen 1971-2000']]
heisz_df.columns = ['Klimakennwert','Name','AGS','Median','Einheit','1971-2000']
heisz_df
```

```python
heisz_df.describe()
```

```python
#split multiple ags numbers into list
heisz_df['AGS'] = heisz_df['AGS'].str[:5]
heisz_df
```

```python
heisz_df['2036-2065'] = heisz_df['1971-2000']+heisz_df['Median']
heisz_df['percentage change'] = heisz_df['Median']/heisz_df['1971-2000']*100
heisz_df
```

```python
#add Bundesland
heisz_df['Bundesland'] = heisz_df['AGS'].str[:2].astype(int).map(regions)
heisz_df
```

```python
heisz_df['Name'] = heisz_df['Name'].str.replace('Region um','').str.strip()
heisz_df
```

```python
heisz_df['Postcard Intro'] = heisz_df.agg(lambda x: f"Ich schicke dir Grüße aus dem schönen {x['Name']} in {x['Bundesland']}! Momentan wird es hier an {x['1971-2000']} Tagen im Jahr über 30 Grad, aber wenn wir so weitermachen, könnten es Mitte des Jahrhunderts schon {x['2036-2065']} sein. {round(x['percentage change'])} Prozent mehr!", axis=1)
```
