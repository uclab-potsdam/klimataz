```python
df = pd.read_csv("collected_gerics_budesland_heisze_tage_raw.csv")
df
```

```python
bl_df = df.copy()
bl_df
```

```python
bl_df['2036-2065'] = bl_df['1971-2000']+bl_df['Median']
bl_df['percentage change'] = bl_df['Median']/bl_df['1971-2000']*100
bl_df
```

```python
bl_df['Postcard Intro'] = bl_df.agg(lambda x: f"Ich schicke dir Grüße aus dem schönen {x['Name']}! Momentan wird es hier an durchschnittlich {str(round(x['1971-2000'],1)).replace('.',',')} Tagen im Jahr über 30 Grad, aber wenn wir so weitermachen, könnten es Mitte des Jahrhunderts schon {str(round(x['2036-2065'],1)).replace('.',',')} sein. {round(x['percentage change'])} Prozent mehr!", axis=1)
```
