```python
# import necessary libraries

import pandas as pd
import numpy as np
```

```python
#load dataset from local file
source_df = pd.read_csv("data/32121-01-02-4-B_flat_haushaltsabfaelle.csv",delimiter=";", encoding='latin1', low_memory=False)

#remove subgroups of waste Haus- und Sperrmüll, Getrennt erfasste organische Abfälle, Getrennt erfasste Wertstoffe, Sonstige Abfälle, Biotonne, 
#remove subgroups of waste Haus- und Sperrmüll, Getrennt erfasste organische Abfälle, Getrennt erfasste Wertstoffe, Sonstige Abfälle and Insgesamt and Insgesamt
source_df = source_df[(source_df['2_Auspraegung_Code'] != 'ABFALLART100') & (source_df['2_Auspraegung_Code'] != 'ABFALLART300') & (source_df['2_Auspraegung_Code'] != 'ABFALLART400') & (source_df['2_Auspraegung_Code'] != 'ABFALLART201') & (source_df['2_Auspraegung_Code'] != 'ABFALLART202') & (source_df['2_Auspraegung_Label'] != "Insgesamt")]

#remove subgroups of waste Haus- und Sperrmüll, Getrennt erfasste organische Abfälle, Getrennt erfasste Wertstoffe, Sonstige Abfälle and Insgesamt
source_df = source_df[(source_df['2_Auspraegung_Code'] != 'ABFALLART100') & (source_df['2_Auspraegung_Code'] != 'ABFALLART300') & (source_df['2_Auspraegung_Code'] != 'ABFALLART400') & (source_df['2_Auspraegung_Label'] != "Insgesamt")]

source_df
```

```python
waste_df = source_df.copy()

#extract year from date
waste_df['Jahr']=waste_df['Zeit'].astype(str)

#drop unnecessary columns
waste_df.drop(columns=['Zeit','Statistik_Code', 'Statistik_Label', 'Zeit_Code', 'Zeit_Label', 
       '1_Merkmal_Code', '1_Merkmal_Label', 
       '2_Merkmal_Code', '2_Merkmal_Label',
       '2_Auspraegung_Code'],inplace=True)

#rename columns
waste_df.rename(columns={'1_Auspraegung_Code':'AGS','1_Auspraegung_Label':'Name','2_Auspraegung_Label':'Abfall','AEW009__Aufkommen_an_Haushaltsabfaellen_(o.E-altgeraete)__t':'Anzahl'},inplace=True)

#replace '-' with 0 and '.' with 'NaN according to Zeichnerklärung
waste_df['Anzahl'].replace('-','0',inplace=True)
waste_df['Anzahl'] = waste_df['Anzahl'].apply(pd.to_numeric,errors='coerce')

waste_df['AGS']=waste_df['AGS'].replace("DG",0)

waste_df
```

```python
#pivot dataframe to show waste and years as columns
pt_df = waste_df.pivot(index=['AGS','Name'],columns=['Abfall','Jahr'],values=['Anzahl'])

#rename columns
pt_df.rename(columns={'Getrennt erfasste organische Abfälle':'organicwaste'},inplace=True)

pt_df
```

```python
#collapse levels of column names and join waste type to year
pt_df.columns = pt_df.columns.map('_'.join)

#sort columns so indicators are together
pt_df.columns=sorted(pt_df.columns)

#prepend all columns with indicator name
pt_df=pt_df.add_prefix('waste_count_')

#convert AGS to numeric and set as new index
pt_df.reset_index(inplace=True)
pt_df['AGS'] = pd.to_numeric(pt_df['AGS'])
pt_df.set_index('AGS',inplace=True)
pt_df.sort_index(ascending=True,inplace=True)

#strip whitespace from region names
pt_df['Name']=pt_df['Name'].str.strip()

pt_df
```

```python
#save dataframe to file
pt_df.to_csv("data/final_data_landkreis_waste_organicwaste.csv")
```
