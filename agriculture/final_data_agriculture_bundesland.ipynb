```python
# import necessary libraries

import pandas as pd
import numpy as np
```

```python
#list of regions

regions = {"Deutschland":0,
           "Baden-Württemberg":8,
           "Bayern":9,"Berlin":11,
           "Brandenburg":12,
           "Bremen":4,
           "Hamburg":2,
           "Hessen":6,
           "Mecklenburg-Vorpommern":13,
           "Niedersachsen":3,
           "Nordrhein-Westfalen":5,
           "Rheinland-Pfalz":7,"Saarland":10,"Sachsen":14,"Sachsen-Anhalt":15,"Schleswig-Holstein":1,"Thüringen":16}

regions

```

```python
#load dataset from local file
source_df = pd.read_csv("data/2207_viehbestand_besatzdichte_bundeslaender_2013_bis_2020.csv",delimiter=",")#, encoding='latin1', low_memory=False)

source_df
```

```python
bd_df = source_df.copy()

#rename unnamed column
bd_df.rename(columns={'Unnamed: 0':'Name'},inplace=True)

bd_df.columns=bd_df.columns.str.replace(" ","_")
bd_df.columns=bd_df.columns.str.replace(",","_")
bd_df.columns=bd_df.columns.str.replace("/","_")

bd_df
```

```python
#calculate fraction of animals living under the 2,0 GV/ha threshhold
bd_df['davon_auf_unter_2_0_GV_ha'] = bd_df['Unter_1_0_GV_ha']+bd_df['1_0-1_5_GV_ha']+bd_df['1_5-2_0_GV_ha']
bd_df
```

```python
#change year column to string to facilitate renaming columns after second pivot
bd_df['Jahr'] = bd_df['Jahr'].astype(str)

#pivot column to have regions as rows and indicators and years as columns
pt_df = bd_df.pivot(index=['Name'],columns=['Jahr'],values=['Insgesamt','davon_auf_unter_2_0_GV_ha', 'davon_Ökolandbau', 'Unter_1_0_GV_ha',
       '1_0-1_5_GV_ha', '1_5-2_0_GV_ha', '2_0-2_5_GV_ha', '2_5-5_0_GV_ha',
       '5_0_und_mehr_GV_ha'])

#collapse levels of column names and join energy type to year
pt_df.columns = pt_df.columns.map('_'.join)

#prepend columns with indicator name
pt_df=pt_df.add_prefix('animalcount_bydensity_')

#add AGS ids from the regions dictionary using the region names
pt_df['AGS']=pt_df.index.map(regions)

pt_df
```

```python
#set AGS as index and copy region names to first column
first_column=pt_df.index
pt_df.set_index('AGS', inplace=True)
pt_df.insert(0,'Name',first_column)

#sort the dataframe by AGS id
pt_df.sort_index(inplace=True)
```
